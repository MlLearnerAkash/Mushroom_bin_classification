```python
from sklearn import preprocessing 
  
# label_encoder object knows how to understand word labels. 
l_encd = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
ms['class']= l_encd.fit_transform(ms['class']) 
  
ms['class'].unique() 
```

**For label-encoding**:we first created a list of the columns in ***'col'***

```python
col=ms.columns

for i in col:
  ms[i]=l_encd.fit_transform(ms[i])  #".fit_transform" label encoding each features column-wise
```

From the heat-map,it is clear that there exists **Multi-colinearity** between some features(viz. the correlation between 'gill-attachment' and 'veil-color' is about 0.9!!).The disadvantage of the **Multi-colinearity**,it makes the model over fitting.

To **overcome** this ambiguity,we will creat a new features out of these colineared features and observe the depenceny on the target features for the new obtained features..

##***Adding New Features***

Multi-colinearity:Although there is no exclusive dependence on any feature,but the internal features depends on each other,that is called Internal Colinearity

```python
ms.columns
```

```python
ms.columns
```

```python
ms['cap_clr+slk_shape']=ms['cap-color']*ms['stalk-shape'] #+0.7
```

```python



#ms['odr+spore_sprint_color']=ms['odor']+ms['spore-print-color']
ms['gill_viel']=ms['gill-attachment']*ms['veil-color']#0.9
ms['gill_rint']=ms['gill-size']*ms['spore-print-color']# 0.6
ms['brs+ring_type']=ms['bruises']*ms['ring-type'] #0.7
```

```python
ms.columns
```

So,We have added **4 more features**: 

1.   **'cap_clr+slk_shape'** :out of **'cap-color'** and **'stalk-shape'**
2.  **'gill_viel'** :out of **'gill-attachment'** and **'veil-color'**
3.  **'gill_rint'**: out of **'gill-size'** and **'spore-print-color'**
4.  **'brs+ring_type'**:out of **bruises** and **'ring-type'**

```python

X_train = w_o_n_f_ms.drop("class", axis=1) #Input-features
Y_train = w_o_n_f_ms["class"] #output/target feature
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.3)# "test_size=0.3"=>30% of the whole data goes to 'x_test' and 'y_test'


lr = LogisticRegression(max_iter=1000) # max_iter=>maximum number of epochs
lr = lr.fit(x_train,y_train)
act = accuracy_score(y_train,lr.predict(x_train))
print('Train Accuracy is: ',(act*100))
```

```python
act2 = accuracy_score(y_test,lr.predict(x_test)) # accuracy for test data 
print('Test Accuracy is: ',(act2*100))
```
